10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:In Python, `map()`, `filter()`, and `reduce()` are built-in functions used to process collections like lists. Each of them serves a different purpose:

 `map()`

 The `map()` function is used to apply a function to every item in a list (or other iterable).
 It transforms each item based on the logic you give.

Example:

python
nums = [1, 2, 3]
result = list(map(lambda x: x * 2, nums))  # [2, 4, 6]


`filter()`

 The `filter()` function is used to select items from a list based on a condition.
 Only the items that satisfy the condition will be kept.

Example:

python
nums = [1, 2, 3, 4]
result = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]

`reduce()`

 The `reduce()` function (from the `functools` module) is used to combine all items in a list into a single value.
 It applies a function cumulatively: first to the first two items, then to the result and the next item, and so on.

Example:

python
from functools import reduce
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums).

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.


```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Take input from user
user_input = input("Enter numbers separated by spaces: ")
number_list = list(map(int, user_input.split()))

# Call the function and print result
result = sum_even_numbers(number_list)
print("Sum of even numbers:", result)
```

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square_numbers(numbers):
    squared = []
    for num in numbers:
        squared.append(num * num)
    return squared

# Take input from user
user_input = input("Enter integers separated by spaces: ")
number_list = list(map(int, user_input.split()))

# Call the function and print result
squared_list = square_numbers(number_list)
print("List of squares:", squared_list)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Take input from user
num = int(input("Enter a number between 1 and 200: "))

# Check and print result
if 1 <= num <= 200:
    if is_prime(num):
        print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
else:
    print("Please enter a number between 1 and 200.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a

# Take input from user
user_input = input("Enter the number of Fibonacci terms: ")
terms = int(user_input.split()[0])


# Create and use the iterator
fib = Fibonacci(terms)
print("Fibonacci sequence:")
for num in fib:
    print(num, end=' ')
print()
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(limit):
    for i in range(limit + 1):
        yield 2 ** i

# Get input from user
num = int(input("Enter the maximum exponent: "))

# Use the generator
print("Powers of 2:")
for value in powers_of_two(num):
    print(value, end=' ')
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # remove newline characters

# Example usage
file_name = "sample.txt"

print("File content:")
for line in read_file_lines(file_name):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def c_to_f(celsius):
    return (celsius * 9/5) + 32

# Get Celsius temperatures from user
celsius_list = input("Enter temperatures in Celsius (separated by spaces): ")
celsius_values = list(map(float, celsius_list.split()))

# Convert using map()
fahrenheit_values = list(map(c_to_f, celsius_values))

# Print result
print("Temperatures in Fahrenheit:", fahrenheit_values)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Function to check if a character is NOT a vowel
def not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Get input from user
text = input("Enter a string: ")

# Use filter to remove vowels
no_vowels = ''.join(filter(not_vowel, text))

# Print result
print("String without vowels:", no_vowels)
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
# List of orders: [Order Number, Book Title, Quantity, Price per Item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Calculate the total price for each order
# Add 10 € if total price is less than 100 €
result = list(map(lambda order: (
    order[0],  # Order Number
    order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0)
), orders))

# Display the output in a user-friendly format
print("Order Summary:")
print("---------------------------")
for order_num, total_price in result:
    print(f"Order Number: {order_num}  -->  Total Price: €{total_price:.2f}")
```
